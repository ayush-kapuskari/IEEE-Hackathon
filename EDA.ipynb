```python
zero_counts = {}

# Counter for columns with at least one zero value
columns_with_zeros = 0

# Iterate over each column in the DataFrame
for column in df.columns:
    # Count the number of zero values in the column
    count = (df[column] == 0).sum()
    
    # If the count is greater than 0, store it and increment the counter
    if count > 0:
        zero_counts[column] = count
        columns_with_zeros += 1

print(f"Total number of columns with zero values: {columns_with_zeros}\n")
print("Number of zero values in each column:")
for column, count in zero_counts.items():
    print(f"  {column}: {count}")
```

## set the thresold value

```python
def correlation(data,thresold):
    corr_set=set()
    corr_matrix=data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i][j]>thresold):
                colname=corr_matrix.columns[i]
                corr_set.add(colname)
    return corr_set       
```

```python
colums_remove=correlation(df,0.9)
```

```python
df.drop(colums_remove,axis=1,inplace=True)
```
